```python
import pandas
import matplotlib.pyplot as plt
from glob import glob

files = glob('../Files/at2019qiz_charlotte/*')
files
```

```python
z = 0.01513

for file in files:
    data = pandas.read_csv(file, sep='\s+', header=None)
    data.columns = ['Wavelength', 'Flux Density', 'Error']
    data.head()

    wl = data['Wavelength']
    flux = data['Flux Density']
    err = data['Error']

    plt.figure()
    plt.axvline(6563, label=r'H$\alpha$', color='red'); plt.axvline(4861, label=r'H$\beta$', color='cyan')
    plt.plot(wl/(1+z), flux, label='data')
    plt.xlabel('$\lambda$ ($\AA$)'); plt.ylabel(r'$f_{\lambda}$ ($erg/s/cm^{2}/\AA$)')
    plt.xlim(6300, 6700)
    plt.title(f'{file}'); plt.legend(); plt.show()
```
