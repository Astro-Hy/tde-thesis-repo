```python
import numpy as np
import matplotlib.pyplot as plt

def power_law(wl, z, A, k, C):
    return A*(wl/(1+z))**k+C
```

```python
wl1, flux1 = np.loadtxt('../Files/at2020zso_efosc2_charlotte/tAT2020zso_20201208_Gr11_Free_slit1.0_59229_1_err.asci', usecols=(0, 1), unpack=True)
wl2, flux2 = np.loadtxt('../Files/ASASSN-18pg_decam_charlotte/a07.txt', usecols=(2, 3), unpack=True)
wl3, flux3 = np.loadtxt('../Files/at2019qiz_charlotte/20210208_MUSE.txt', usecols=(0, 1), unpack=True)
wl4, flux4 = np.loadtxt('hyzspec.txt', unpack=True)

plt.figure(figsize=(10, 5))
plt.plot(wl4/(1+0.04573), 0.3*(flux4 - power_law(wl4, 0.04573, *[ 1.03658933e-13, -1.93697829e-03, -1.01611922e-13]) + 5e-16), label='AT 2018hyz')
plt.plot(wl2, 0.1*(flux2 - power_law(wl2, 0, *[-5.69314816e-12, -1.78149645e-04,  5.68527108e-12])), label='AT 2018dyb')
plt.plot(wl3/(1+0.01513), 0.1*(flux3 - power_law(wl3, 0.01513, *[-2.57426216e-06, -4.41129826e+00,  7.68577702e-16]) - 18e-16), label='AT 2019qiz')
plt.plot(wl1/(1+0.0563), 2*(flux1 - power_law(wl1, 0.0563, *[ 8.74764134e-08, -2.43552278e+00,  7.81167947e-17])) - 3.7e-16, label='AT 2020zso')
plt.axvline(6563.28, linestyle='--', color='black', linewidth=0.5, label=r'H$\alpha$')
plt.xlabel(r'Rest $\lambda$ ($\AA$)'); plt.legend()
plt.yticks([]); plt.ylabel(r'f$_\lambda$ scaled + constant')
plt.savefig('comp.png')
plt.show()
```
