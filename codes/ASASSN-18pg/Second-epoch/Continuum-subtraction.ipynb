```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

def power_law(wl, A, k, C):
    return A*(wl/(1+z))**k+C

def graph_more():
    plt.axvline(6563, label=r'H$\alpha$', color='red')
    plt.axvline(4861, label=r'H$\beta$', color='cyan')
    plt.xlabel(r'Rest $\lambda$ ($\AA$)'); plt.legend(loc=1); plt.show()

spec = np.loadtxt('spec.txt')
wl = spec[:, 0]
flux = spec[:, 1]
z = 0.017932
```

```python
plt.figure(figsize=(12, 5))
plt.plot(wl/(1+z), flux + 3e-17, label='data')
plt.ylabel(r'$f_{\lambda}$')
plt.title('AT2018dyb Spectrum')
plt.axhline(0, ls='--', c='k')
graph_more()
```
