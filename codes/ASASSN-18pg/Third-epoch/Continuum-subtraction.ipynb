```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

def power_law(wl, A, k, C):
    return A*(wl/(1+z))**k+C

def graph_more():
    plt.axvline(6563, label=r'H$\alpha$', linewidth=0.5, linestyle='--', color='red')
    # plt.axvline(4861, label=r'H$\beta$', color='cyan')
    plt.xlabel(r'Rest $\lambda$ ($\AA$)'); plt.ylabel(r'f$_\lambda$ ' + ' (10$^{17}\cdot$erg/s/cm$^2$/Hz)')
    plt.legend(loc=1); plt.show()

spec = np.loadtxt('third-spec.txt')
spectrum = spec[~np.isnan(spec).any(axis=1)]
wl = spectrum[:, 0]
flux = spectrum[:, 1]
z = 0.017932
```

```python
continuum = curve_fit(power_law, wl, flux, maxfev=10000)
pl_params = continuum[0]
```

```python
plt.figure(figsize=(12, 5))
plt.plot(wl/(1+z), flux, label='data')
plt.plot(wl/(1+z), power_law(wl, *pl_params), label='fit')
plt.ylabel(r'$f_{\lambda}$')
plt.title('AT2018dyb Spectrum')
graph_more()
```

```python
wl_3 = np.argmin(np.abs(wl/(1+z) - 5000)); wl_4 = np.argmin(np.abs(wl/(1+z) - 6400))
wl_5 = np.argmin(np.abs(wl/(1+z) - 6700)); wl_6 = np.argmin(np.abs(wl/(1+z) - 7200))
cut_range = np.concatenate((np.arange(0, wl_3), np.arange(wl_4, wl_5), np.arange(wl_6, len(wl)-1)))
wl_cut = np.delete(wl, cut_range); flux_cut = np.delete(flux, cut_range)
```

```python
continuum = curve_fit(power_law, wl_cut, flux_cut, p0=pl_params, maxfev=int(1e5))
#continuum = [-4.35629287e-12,  1.71834564e-04,  4.36290637e-12]

plt.figure(figsize=(12, 5))
plt.plot(wl/(1+z), flux, label='data'); plt.plot(wl/(1+z), power_law(wl, *continuum[0]), label='fit')
plt.axvspan(wl[0]/(1+z), wl[wl_3]/(1+z), alpha=0.1)
plt.axvspan(wl[wl_4]/(1+z), wl[wl_5]/(1+z), alpha=0.1); plt.axvspan(wl[wl_6]/(1+z), wl_cut[-1]/(1+z), alpha=0.1)
plt.ylabel(r'$f_{\lambda}$'); plt.title('Spectrum with Powerlaw Continuum Fit')
graph_more()

plt.figure(figsize=(12, 5))
plt.axhline(0, color='black', linestyle='dotted')
plt.plot(wl/(1+z), flux - power_law(wl, *continuum[0]), color='black', label='spectrum')
#plt.xlim(6100, 7000); plt.ylim(-3e-16, 5e-16)
plt.title('Spectrum with Subtracted Continuum')
graph_more()
```
